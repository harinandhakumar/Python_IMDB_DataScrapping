```python
import pandas as pd
adventure = pd.read_csv("Adventure.csv")
animation = pd.read_csv("Animation.csv")
biography = pd.read_csv("Biography.csv")
crime = pd.read_csv("Crime.csv")
family = pd.read_csv("Family.csv")
csv_files = ["Adventure.csv", "Animation.csv", "Biography.csv", "Crime.csv", "Family.csv"]

from sqlalchemy import create_engine

# Define database connection URL
db_connection = create_engine("mysql+mysqlconnector://root:root@localhost/python")

# Test connection
conn = db_connection.connect()
print("Connected successfully!")

# Insert Data into MySQL Table
adventure.to_sql("adventure", con=db_connection, if_exists="replace", index=False)
animation.to_sql("animation", con=db_connection, if_exists="replace", index=False)
biography.to_sql("biography", con=db_connection, if_exists="replace", index=False)
crime.to_sql("crime", con=db_connection, if_exists="replace", index=False)
family.to_sql("family", con=db_connection, if_exists="replace", index=False)

print("Individual CSV data inserted into MySQL successfully!")

df_list = [pd.read_csv(file) for file in csv_files]
merged_df = pd.concat(df_list, ignore_index=True)
merged_df.to_csv("merged_file.csv", index=False)

print("Merged CSV saved successfully!")

merged_df.to_sql("merged", con=db_connection, if_exists="replace", index=False)

print("Merged data inserted into MySQL successfully!")

# Close connection
conn.close()
```
